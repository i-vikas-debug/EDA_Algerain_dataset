# Data Cleaning

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df = dataset
```

```python
df.head()
```

```python
df.info()
```

```python
df[['Region']]= df[['Region']].astype(int)
```

```python
df.head()
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
# removing 122nd col bcoz not a valid info
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# removing extra space from head columns name
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df.info()
```

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
objects = [ i for i in df.columns if df[i].dtypes=='O']
```

```python
objects
```

```python
for i in objects:
    if i != 'Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.to_csv('Algerian_forest_fires_dataset_cleaned.csv',index=False)
```
